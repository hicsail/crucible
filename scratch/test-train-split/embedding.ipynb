```python
import json
prettyprint = lambda x: json.dumps(x, indent=4, sort_keys=True)
with open('result.json', 'r') as f:
    aq_plans_json = json.load(f)
```

```python
print(prettyprint(aq_plans_json['plans'][0]))
```

```python
estr = ''
for op in aq_plans_json['plans'][0]['operations']:
    for i in op['inputs']:
        estr += i['sample'] + i['object']['function'] + i['object']['material'] + i['object']['type']
        estr += '^'
    estr += '>'
    if len(op['outputs']) == 0:
        estr += 'end'
    estr += '\n'
    for o in op['outputs']:
        estr += i['sample'] + i['object']['function'] + i['object']['material'] + i['object']['type'] 
print(estr)
```

```python
from gensim.models import Word2Vec
model = Word2Vec(estr, 
                 min_count=3,   # Ignore words that appear less than this
                 size=200,      # Dimensionality of word embeddings
                 workers=2,     # Number of processors (parallelisation)
                 window=5,      # Context window for words during training
                 iter=30)  
```
